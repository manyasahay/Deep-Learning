Comparing learning rates

```python
import matplotlib.pyplot as plt
from keras.optimizers import Adam

def fit_and_plot_learning_rates(learning_rates, X_train, y_train, X_val, y_val, epochs=50, batch_size=16):
    histories = []

    for lr in learning_rates:
        opt = Adam(learning_rate=lr)
        model = Sequential()
        model.add(Dense(32, activation='relu', input_shape=(6,)))
        model.add(Dense(32, activation='relu'))
        model.add(Dense(16, activation='relu'))
        model.add(Dense(1, activation='sigmoid'))
        model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
        history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_val, y_val))
        histories.append(history)

    # Plot learning curves for each learning rate
    for i, lr in enumerate(learning_rates):
        plt.plot(histories[i].history['val_loss'], label=f'LR={lr}')

    plt.title('Validation Loss for Different Learning Rates')
    plt.xlabel('Epochs')
    plt.ylabel('Validation Loss')
    plt.legend()
    plt.show()


learning_rates = [0.1, 0.01, 0.001]
fit_and_plot_learning_rates(learning_rates, X_train, y_train, X_val, y_val)
```

```python
import matplotlib.pyplot as plt
from keras.optimizers import SGD
from keras.models import Sequential
from keras.layers import Dense

def fit_and_plot_momentums(momentums, X_train, y_train, X_val, y_val, learning_rate=0.01, epochs=50, batch_size=16):
    histories = []

    for momentum in momentums:
        opt = SGD(learning_rate=learning_rate, momentum=momentum)
        model = Sequential()
        model.add(Dense(32, activation='relu', input_shape=(6,)))
        model.add(Dense(32, activation='relu'))
        model.add(Dense(16, activation='relu'))
        model.add(Dense(1, activation='sigmoid'))
        model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
        history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_val, y_val))
        histories.append(history)

    # Plot learning curves for each momentum value
    for i, momentum in enumerate(momentums):
        plt.plot(histories[i].history['val_loss'], label=f'Momentum={momentum}')

    plt.title('Validation Loss for Different Momentum Values')
    plt.xlabel('Epochs')
    plt.ylabel('Validation Loss')
    plt.legend()
    plt.show()


momentums = [0.0, 0.5, 0.9]
fit_and_plot_momentums(momentums, X_train, y_train, X_val, y_val, learning_rate=0.01)
```

Accuracy for momentum = 0.0 (82.09)

momentum = 0.5 (81.34)

momentum = 0.9 (79.85)

All values in brackets are the validation accuracy.

```python
def fit_and_plot_learning_rate_schedules(learning_rate_schedules, X_train, y_train, X_val, y_val, momentum=0.0, epochs=20, batch_size=16):
    histories = []

    for lr_schedule in learning_rate_schedules:
        opt = SGD(learning_rate=lr_schedule, momentum=momentum)
        model = Sequential()
        model.add(Dense(32, activation='relu', input_shape=(6,)))
        model.add(Dense(32, activation='relu'))
        model.add(Dense(16, activation='relu'))
        model.add(Dense(1, activation='sigmoid'))
        model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
        history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_val, y_val))
        histories.append(history)

    # Plot learning curves for each learning rate schedule
    for i, lr_schedule in enumerate(learning_rate_schedules):
        plt.plot(histories[i].history['val_loss'], label=f'LR Schedule={lr_schedule}')

    plt.title('Validation Loss for Different Learning Rate Schedules')
    plt.xlabel('Epochs')
    plt.ylabel('Validation Loss')
    plt.legend()
    plt.show()

learning_rate_schedules = [0.01, 0.001, 0.0001]
fit_and_plot_learning_rate_schedules(learning_rate_schedules, X_train, y_train, X_val, y_val, momentum=0.0)
```

0.01 is the best learning rate (80.97 validation accuracy)

```python

from tensorflow.keras.optimizers import Adam, Adagrad, RMSprop

def fit_and_plot_optimizers(optimizers, X_train, y_train, X_val, y_val, epochs=20, batch_size=16):
    histories = []

    for optimizer in optimizers:
        model = Sequential()
        model.add(Dense(32, activation='relu', input_shape=(6,)))
        model.add(Dense(32, activation='relu'))
        model.add(Dense(16, activation='relu'))
        model.add(Dense(1, activation='sigmoid'))
        model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
        history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_val, y_val))
        histories.append(history)

    # Plot learning curves for each optimizer
    optimizer_names = [optimizer.__class__.__name__ for optimizer in optimizers]

    for i, optimizer_name in enumerate(optimizer_names):
        plt.plot(histories[i].history['val_loss'], label=optimizer_name)

    plt.title('Validation Loss for Different Optimizers')
    plt.xlabel('Epochs')
    plt.ylabel('Validation Loss')
    plt.legend()
    plt.show()


adam_optimizer = Adam(learning_rate=0.01)
adagrad_optimizer = Adagrad(learning_rate=0.01)
rmsprop_optimizer = RMSprop(learning_rate=0.01)

optimizers = [adam_optimizer, adagrad_optimizer, rmsprop_optimizer]
fit_and_plot_optimizers(optimizers, X_train, y_train, X_val, y_val)
```

Best optimizer would be either RMSprop(82.46) or Adam (80.60) as we can see from this graph. RMSProp has a high validation accuracy.
